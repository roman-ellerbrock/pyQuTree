```python
from qutree import *

def double_well_2(x):
    # p = [1, 2, 3, ...]
    point = np.array(list(range(x.shape[0])))
    return np.sum((x-point)**2)

N = 21
r = 4
f = 3

objective = Objective(double_well_2)

# G = balanced_tree(f, r, N)
# plot_tree(G)

G = tensor_train_graph(f, r, N)
fig = plot_tt_diagram(G)

primitive_grid = [np.linspace(0., 4., num = N)] * f

G = tn_grid(G, primitive_grid)

G_opt, df = ttnopt(G, objective, nsweep = 3)

print(df.loc[df["f"].idxmin()])
print(f"Number of objective function calls: {objective.function_calls}")
print(f"Number of cached function accesses: {objective.cache_hits}")
display(df)
```
